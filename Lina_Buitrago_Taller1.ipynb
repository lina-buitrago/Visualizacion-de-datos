### **Visualización uno:**

>**U.S. House Election Results 2018**  
la información de la primera visualización fue tomada del NYTimes, acerca de los resultados a las elecciones a la cámara del senado en Estados Unidos del año 2018, dando como contexto la victoria de los demócratas ante los republicanos y liberales, así mismo representa la cantidad y proporción de votos obtenida en cada distrito de Estados Unidos.


![Elecciones](https://drive.google.com/uc?export=view&id=1CGYyYTaHYYWH_6rtf2EjVm58soCbWRhh) 

>**¿Qué?**  

Los datos que quiere presentar la visualización es la cantidad de votos y proporción d ellos mismos obtenidas en las elecciones a la cámara del senado 2018 entre Demócratas, republicanos liberales en cada distrito de los Estados Unidos, podría decirse que los datos son categóricos ordenados por ubicación geográfica, ya que, dependiendo de la predominancia de voto de cada partido, se asigna un color que los distingue e identifica entre sí.  

>**¿Por qué?**  

+	Con esta visualización se pretende **identificar** y **comparar** la preferencia política entre demócratas, Republicanos y liberales para el senado a la cámara para el año 2018, en cada distrito de Estados Unidos. Obteniendo como resultado visual el **resumen** de los votos obtenidos y la victoria de los demócratas.  

+ Si bien a primera vista se observa en la mayoría de distritos una tendencia en la preferencia hacia los republicanos, la interactividad de la visualización permite observar la cantidad y proporción de votos en cada distrito de manera más detallada y adicionalmente proporciona la visualización **cartográfica** de los votos, donde se observa mayormente el voto hacia los demócratas.  

![Cartografico](https://drive.google.com/uc?export=view&id=1UgFaiZg8KEwh0DIDyTNHFiaOHThi4hGK) 

>**¿Cómo?**  

+	La visualización cuenta con una interacción sencilla, presentando el mapa político de Estados Unidos, que, al momento de sobreponer el cursor en cada distrito, presenta la cantidad de votos obtenida en dicho distrito para cada una de las ideologías políticas (diferenciadas entre si por un color que las caracteriza) y la proporción de los votos.  

+	La visualización también permite observar con detenimiento la distribución geográfica de los votos en cada distrito al dar click en él.  

+	Los pasos para esta visualización fueron codificar -> Analizar y yuxtaponer.  

>**Marcas usadas:**  

+ Marca de área geográfica.  

>**Canales usados:**  

+ Canal de color para cada atributo categórico el cual nos permite distinguir la preferencia de cada ideología política de forma geográfica.

+ Canal de popout por color, para distinguir los distritos donde se presenta un ´giro´ (es decir, el distrito voto por un partido diferente al del año pasado)   

>**Cumple o no cumple alguna de las reglas generales descritas:**  

La visualización si cumple con las reglas generales, pues utiliza la interacción en la imagen. Resulta atractiva visualmente, explicando bien la distribución de los votos y detallado la proporción de estos en que cada estado y para cada categoría. adicionalmente muestra la comparación de cada categoría que cumple con una paleta de colores que logra diferenciar a cada una.  

>**Mejoras posibles:**  

La visualización está bien construida, se logra ver la distinción de cada partido y la interactividad en la misma permite disponer de información más detallada, donde se logra observar con mayor detenimiento los colores y el significado de los mismos. Una posible mejora podría ser marcar con un tono más pronunciado los rebordes de los diminutivos de cada estado, ya que en algunos donde predomina el color azul no se distingue tan bien.

### **Visualización Dos:**  

>**Coranavirus Riskiest activities**  

La información de la segunda visualización fue tomada de diversas fuentes tales como: NYTimes, Reuters, NPR, SF Gate y otros, siendo la web ‘Information is beautiful’ la creadora de la visualización. En esta visualización se observa el nivel de riesgo de contagio de coronavirus, que tienen actividades cotididianas, según diversos epidemiólogos y profesionales en salud.   


![Coronavirus](https://drive.google.com/uc?export=view&id=1Log84bAiifJH923OBLjCsorGKbrqw7S7) 

>**¿Qué?**  

Los datos que quiere presentar esta visualización son **categóricos** y **ordinales secuenciales**, ya que presenta **atributos** y/o factores que determinan el nivel de riesgo son: la cantidad de personas, que tan amplio o reducido es el espacio, cuanto tiempo se estará en dicho espacio, la localización del espacio (es al aire libre o cerrado), superficies con alto contacto, el número de casos en el área y que tan probable es que se cumpla con medidas sanitarias de seguridad.  

>**¿Porqué?**  

Con esa visualización se busca **analizar** y **presentar** el nivel de riesgo de contagio al que estamos expuestos al desarrollar cierto tipo de actividades, teniendo en cuenta diversos **atributos** y/o factores que **influyen** o se relacionan directamente con el aumento o disminución del riesgo de contangio.

>**¿Cómo?**  

+ Se observa una visualización no interactiva de un gráfico de burbujas, ordenado y secuencial, donde se hace uso de la saturación de colores, tamaños y ángulos para representar el aumento o disminución del riesgo de contagio.  

+ Los pasos de esta visualización fueron: codificar -> navegar -> seleccionar -> filtrar -> yuxtaponer -> ordenar

>**Marcas usadas:**   

+ Marca de puntos.

>**Canales usados:**  

+ Canal espacial tanto vertical como horizontal.  
+ Canal de saturación de color.  
+ Canal de área. 
+ Canal de tamaño.  

>**Cumple o no cumple alguna de las reglas generales descritas:**   

La visualización cumple con las reglas generales descritas, presenta la información necesaria para el entendimiento del lector, explicando el nivel de riesgo de cada una de las actividades presentadas y resaltando (aunque de manera sutil) ambos extremos (el riego leve y alto).  

>**Mejoras posibles:**  

Si bien en la visualización se mencionan los factores que determinan el riesgo, verticalmente no es muy claro si la posición de las actividades se ubica de esa forma por algún factor en específico, adicionalmente si bien se entiende la intención de la gama de colores, es algo sutil, lo que podría no notarse por algunas personas, por ultimo la letra dentro de las burbujas puede verse algo amontonada.  

**Fuente:** https://informationisbeautiful.net/visualizations/covid-19-coronavirus-infographic-datapack/#treatments
